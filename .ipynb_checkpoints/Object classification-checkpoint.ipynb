```python
import pandas as pd
import numpy as np
import cv2
import os
import tensorflow as tf
import os

from tensorflow.keras import layers
import matplotlib.pyplot as plt

from tensorflow.data import Dataset
import matplotlib.pyplot as plt
```

```python
df_train  = pd.read_csv('train.csv')
```

```python
df_train.head()
```

```python
image_id = df_train['image_id']

labels = df_train['healthy']
```

```python
source_path = os.path.join('images')
dec_path = os.path.join('img')
```

for train , clas in zip(image_id,labels):

```python
image_width = 2048
image_height = 1365
channel = 3
Batch_size = 12
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
directory='img',
    shuffle=True,
    image_size=(image_width,image_height),
    batch_size = Batch_size
    
)
```

```python
n_classes = dataset.class_names
```

```python
plt.figure(figsize=(10,15))
for images,labels in dataset.take(1):
    for i in range(12):
        plt.subplot(3,4,i+1)
        plt.title(f'labels = {labels[i].numpy()}')
        plt.imshow(images[i].numpy().astype('uint8'))
        plt.axis('off')
```

### Spliting data into train,test,validation

```python
len(dataset)
```

```python
def get_data_partition(ds,train_split=0.8,val_split=0.1,test_split=0.1):

  ds_size = len(ds)
  
  train_size = int(ds_size*train_split)
  
  val_size = int(ds_size*val_split)
  
  train_ds = ds.take(train_size)
  
  val_ds = ds.skip(train_size).take(val_size)
  
  test_ds = ds.skip(train_size).skip(val_size)
  
  
  return train_ds,test_ds,val_ds
```

```python
train_ds,test_ds,val_ds = get_data_partition(dataset)
```

```python
print(len(train_ds))
print(len(test_ds))
print(len(val_ds))
```

```python
image_width = 120
image_height = 120
```

```python
 model = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(image_width,image_height),
    layers.experimental.preprocessing.Rescaling(1.0/255),
     
    layers.Conv2D(32,(3,3),activation='ReLU',input_shape=(Batch_size,image_width,image_height,channel)),
    layers.MaxPooling2D(pool_size=(3,3)),
    layers.Dropout(0.25),
     
    layers.Conv2D(64,(3,3),activation='ReLU'),
    layers.MaxPooling2D(pool_size=(3,3)),
    layers.Dropout(0.25),

    layers.Conv2D(32,(3,3),activation='ReLU'),
    layers.MaxPooling2D(pool_size=(3,3)),
     
    layers.Flatten(),
    layers.Dense(2,activation='softmax')
 ])
```

```python
model.build(input_shape=(Batch_size,image_width,image_height,channel))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),metrics='accuracy')
```

```python
del images
del labels
```

```python
model_history = model.fit(train_ds,batch_size=Batch_size,validation_data=val_ds,epochs=35)
```

```python
model.save('model_best')
```

```python
model.evaluate(test_ds)
```

```python
model_history.params
```

```python
plt.plot(model_history.history['loss'],label='loss')
plt.plot(model_history.history['val_loss'],label='val_loss')
plt.legend()
plt.grid()
```

```python
plt.plot(model_history.history['accuracy'],label='loss')
plt.plot(model_history.history['val_accuracy'],label='val_loss')
plt.legend()
plt.grid()
```

```python
y_pred = model.predict(test_ds)
```

```python
y_pred = []
y_test = []
for image,label in test_ds.as_numpy_iterator():
    pred = model.predict(image)
    for index in range(pred.shape[0]):
        y_pred.append(int(n_classes[np.argmax(pred[index])]))
        y_test.append(int(label[index]))
```
