```python
import pandas as pd
import numpy as np
import cv2
import os
import tensorflow as tf
import os

import matplotlib.pyplot as plt

from tensorflow.data import Dataset
import matplotlib.pyplot as plt
```

```python
df_train  = pd.read_csv('train.csv')
```

```python
df_train.head()
```

```python
image_id = df_train['image_id']

labels = df_train['healthy']
```

```python
source_path = os.path.join('images')
dec_path = os.path.join('img')
```

```python
for train , clas in zip(image_id,labels):
    with open(os.path.join(source_path,'{}.jpg'.format(train)),'rb') as scr:
            with open(os.path.join(dec_path,str(clas),'{}.jpg'.format(train)),'wb') as dec:
                dec.write(scr.read())
            
```

```python
image_width = 2048
image_height = 1365
channel = 3
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
directory='img',
    shuffle=True,
    image_size=(image_width,image_height)
)
```

```python
dataset.class_names
```

```python
for images,labels in dataset.take(1):
    for i in range(12):
        plt.subplot(3,4,i+1)
        plt.title(labels[i].numpy())
        plt.imshow(images[i].numpy().astype('uint8'))
        plt.axis('off')
```

### Spliting data into train,test,validation

```python
len(dataset)
```

```python
def get_data_partition(ds,train_split=0.8,val_split=0.1,test_split=0.1):
    
    ds_size = len(ds)
    
    train_size = int(ds_size*train_split)
    
    val_size = int(ds_size*val_split)
    
    train_ds = ds.take(train_size)
    
    val_ds = ds.skip(train_size).take(val_size)
    
    test_ds = ds.skip(train_size).skip(val_size)
    
    
    return train_ds,test_ds,val_ds
```

```python
train_ds,test_ds,val_ds = get_data_partition(dataset)
```
